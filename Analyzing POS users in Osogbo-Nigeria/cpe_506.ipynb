##Employment Status and Usage of POS##

96.8% of the participants had used the service of a POS but those who had not used it before were unemployed and mostly secondary school certificate holders with an exception of a self-employed BSc holder who is considered an outlier.


```python
fig, (ax1, ax2) = plt.subplots(1,2)
fig.suptitle('Education and Employment of Non-POS Users')

x = (none_POS[none_POS["Have you ever used a POS"] == 0])["Level of Education"].value_counts(sort=False).index
y = (none_POS[none_POS["Have you ever used a POS"] == 0])["Level of Education"].value_counts(sort=False).values

x2 = (none_POS[none_POS["Have you ever used a POS"] == 0])["Employment Status"].value_counts(sort=False).index
y2 = (none_POS[none_POS["Have you ever used a POS"] == 0])["Employment Status"].value_counts(sort=False).values

ax1.bar(x, y, color = 'b', width=0.50)
ax1.set_title("Level of Education")
ax1.label_outer()

ax2.bar(x2, y2, color = 'r', width=0.50)
ax2.set_title("Employment Status")
ax2.label_outer()

plt.show()
```

59% of POS users were Female with the remaining 41% being male, this revealed that females were slightly more involved in POS services.
59% of POS users were single, 40% were married and just 1% was divorced, this revealed again a slightly higher usage of POS services by single.
However 55% of POS users were being the age of 21 and 30, this showed very high adoption of POS usage by the youths compared to that of the other age ranges (<20: 9%, 31-40: 26%, 41-50: 9%, 51-60: 2%).
64% of POS users came from a household of a size of 1 to 4 members, this shows a high adoption of POS usage from people come from a small household.

```python
fig, axs = plt.subplots(2,2)
non_cor = ["Gender", "Marital Status", "Age years", "Household Size"]
x = []
y = []
axes = [(0, 0),
       (0, 1,),
       (1, 0),
       (1, 1),]
for idx, col in enumerate(non_cor):
    y_val = (none_POS[col]).value_counts(sort=False).values

    y_per2 = []
    for value in y_val:
        y_per2.append((value / y_val.sum()) * 100)

    x.append((none_POS[col]).value_counts(sort=False).index)
    y.append(y_per2)
    axs[axes[idx]].bar(x[idx], y[idx], width=0.5)
    axs[axes[idx]].set_title(col)

fig.set_figheight(7.5)
fig.set_figwidth(15)
plt.show()
```

According to the data, 46% of businesses are between 2 to 11 years old  while a close amount, 42% are only just 2 years old.
42% of businesses started using POS between 1 to 3 years and 33% of businesses started in less than a year.
This data indicates that the adoption of POS is mostly just recent in Osogbo, the high amount of new businesses seems to indicate that they are the source of the recent influx of POS operations

```python
fig, (ax1, ax2) = plt.subplots(1,2)
fig.suptitle("Age of Business and how long they've been using POS")

x = (POS_user["Age of the business"]).value_counts(sort=False).index
y = (POS_user["Age of the business"]).value_counts(sort=False).values
y_per = []
for value in y:
    y_per.append((value / y.sum()) * 100)

x2 = (POS_user["How many years ago have you been using POS"]).value_counts(sort=False).index
y2 = (POS_user["How many years ago have you been using POS"]).value_counts(sort=False).values
y_per2 = []
for value in y2:
    y_per2.append((value / y2.sum()) * 100)

ax1.bar(x, y_per, color = 'b', width=0.50)
ax1.label_outer()
ax1.set_title("Age of Business")
fig.set_figheight(5)
fig.set_figwidth(10)

ax2.bar(x2, y_per2, color = 'r', width=0.50)
ax2.label_outer()
ax2.set_title("Time since use of POS")

plt.show()
```

55% of POS opeerators have a qualification of HND/B.Sc. and above, 33% are SSCE holders and 14% are OND holders.
